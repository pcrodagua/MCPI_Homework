```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('bike_data.csv')
data = data.drop(['dteday'], axis=1)
data.head()
```

```python
normalized_df=(data-data.mean())/data.std()
train, test = train_test_split(normalized_df, test_size=0.3, random_state=12345)
normalized_df.head()
```

```python


def prot_curve(train_error, test_error):
    plt.plot(np.arange(0,train_error.shape[0]-1), train_error[1:], label = 'train error')
    plt.plot(np.arange(0,train_error.shape[0]-1),test_error[1:], label = 'test error', color= 'green')
    plt.title('Training Set')
    plt.xlabel('Polynomial Complexity (x)')
    plt.ylabel('MSE (y)')
    plt.grid()
    plt.legend()
    plt.show() 

    
def obtain_error(model, classifier, x_train, y_train, x_test, y_test):
    m, n = x_train.shape
    train_error = 0
    test_error = 0
    hyp_model = None
    if classifier:
        cls_ = model.fit(x_train, y_train)
        hyph_train = cls_.predict(x_train)
        hyph_test = cls_.predict(x_test)
        train_error = mean_squared_error(y_train, hyph_train)
        test_error  = mean_squared_error(y_test, hyph_test)
    else:
        reg = LinearRegression().fit(x_train, y_train)
        hyph_train = reg.predict(x_train)
        hyph_test = reg.predict(x_test)
        train_error = mean_squared_error(y_train, hyph_train)
        test_error  = mean_squared_error(y_test, hyph_test)
        
    #prot_(train_error, test_error, n)
    return train_error, test_error


def create_curve():
    nx_train = train.iloc[:,:-1].values
    ny_train = train.iloc[:,-1].values
    # ---- 
    nx_test = test.iloc[:,:-1].values
    ny_test = test.iloc[:,-1].values
    # ------
    train_error = np.zeros(nx_train.shape[1])
    test_error = np.zeros(nx_train.shape[1])

    n = nx_train.shape[1]
    for i in range(1, n):
        train_error[i], test_error[i] = obtain_error(LinearRegression(), False, nx_train[:,0:i], ny_train, nx_test[:,0:i], ny_test)
    prot_curve(train_error, test_error)
```

```python
create_curve()
```
