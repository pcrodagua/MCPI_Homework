```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import mean_squared_error
from sklearn import linear_model

%matplotlib inline
```

```python
data = pd.read_csv('bike_data.csv')
data = data.drop(['dteday'], axis=1)
data.head()
```

```python
normalized_df=(data-data.mean())/data.std()
train, test = train_test_split(normalized_df, test_size=0.3)
normalized_df.head()
```

```python
m, n = train.shape
error_train = np.zeros(n)

nx_train = train.iloc[:,:-1].values
ny_train = train.iloc[:,-1].values
# ---- 
nx_test = test.iloc[:,:-1].values
ny_test = test.iloc[:,-1].values
# ---
nx_train = np.append(np.ones(shape=(nx_train.shape[0],1)),nx_train,axis = 1)
nx_test = np.append(np.ones(shape=(nx_test.shape[0],1)),nx_test,axis = 1)
```

```python
train_error = np.zeros(n)
test_error = np.zeros(n)

theta = np.zeros(n)

for i in range(1, n):
    train_ = nx_train[:,0:i]
    test_ = nx_test[:,0:i]
    
    theta = np.dot(np.linalg.pinv(np.dot(np.transpose(train_), train_)), np.dot(np.transpose(train_), ny_train))
    theta_ = np.dot(np.linalg.pinv(np.dot(np.transpose(test_), test_)), np.dot(np.transpose(test_), ny_test))
    
    hyph_train = np.dot(train_, np.transpose(theta))
    hyph_test = np.dot(test_, np.transpose(theta_))
    
    train_error[i] = mean_squared_error(hyph_train, ny_train)
    test_error[i] = mean_squared_error(hyph_test, ny_test)    
```

```python
plt.plot(np.arange(n-1),train_error[1:], label = 'train error')
plt.plot(np.arange(n-1),test_error[1:], label = 'cross validation error', color= 'green')
plt.title('Training Set')
plt.xlabel('Polynomial Complexity (x)')
plt.ylabel('MSE (y)')
plt.grid()
plt.legend()
plt.show()
```
