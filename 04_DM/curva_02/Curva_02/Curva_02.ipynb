```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
data = pd.read_csv('bike_data.csv')
data = data.drop(['dteday'], axis=1)
data.head()
```

```python
normalized_df=(data-data.mean())/data.std()
train, test = train_test_split(normalized_df, test_size=0.3, random_state=54321)
normalized_df.head()
```

```python
m, n = train.shape
error_train = np.zeros(n)

nx_train = train.iloc[:,:-1].values
ny_train = train.iloc[:,-1].values
# ---- 
nx_test = test.iloc[:,:-1].values
ny_test = test.iloc[:,-1].values


ny_train = ny_train.reshape(-1,1)
ny_test = ny_test.reshape(-1,1)
```

```python
def prot_curve(train_error, test_error, n):
    plt.plot(np.arange(0,n-1), train_error[1:], label = 'train error')
    plt.plot(np.arange(0,n-1),test_error[1:], label = 'cross validation error', color= 'green')
    plt.title('Training Set')
    plt.xlabel('Number Trainning Example (x)')
    plt.ylabel('MSE (y)')
    plt.grid()
    plt.legend()
    plt.show() 

    
def obtain_error(model, classifier, x_train, y_train, x_test, y_test):
    m, n = x_train.shape
    train_error = 0
    test_error = 0
    hyp_model = None
    if classifier:
        cls_ = model.fit(x_train, y_train)
        hyph_train = cls_.predict(x_train)
        hyph_test = cls_.predict(x_test)
        train_error = mean_squared_error(y_train, hyph_test)
        test_error  = mean_squared_error(y_test, hyph_test)
    else:
        reg = model().fit(x_train, y_train)
        hyph_train = reg.predict(x_train)
        hyph_test = reg.predict(x_test)
        train_error = mean_squared_error(y_train, hyph_test)
        test_error  = mean_squared_error(y_test, hyph_test)
    return train_error, test_error


def create_curve():
    nx_train = train.iloc[:,:-1].values
    ny_train = train.iloc[:,-1].values
    # ---- 
    nx_test = test.iloc[:,:-1].values
    ny_test = test.iloc[:,-1].values
    # ------
    train_error = np.zeros(nx_test.shape[0])
    test_error = np.zeros(nx_test.shape[0])

    m = nx_test.shape[0]
    model = LinearRegression()
    for i in range(1, m):
        train_error[i], test_error[i] = obtain_error(model, False, nx_train[0:i,:], ny_train[0:i,], nx_test[0:i,:], ny_test[0:i,])
    prot_curve(train_error, test_error, m)
```
