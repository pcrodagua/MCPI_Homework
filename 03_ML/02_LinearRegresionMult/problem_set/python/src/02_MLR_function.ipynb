 By doing this, we will be able to study the effect of each feature on the dependent variable 

```python
# main
def main():
    '''
        - 01 - instant: record index (DELETED)
        - 02 -dteday : date(DELETED)
        - 03 -season : season (1:springer, 2:summer, 3:fall, 4:winter)
        - 04 -yr : year (0: 2011, 1:2012)
        - 05 -mnth : month ( 1 to 12)
        - 06 -hr : hour (0 to 23)
        - 07 -holiday : weather day is holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
        - 08 -weekday : day of the week
        - 09 -workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
        - 10 weathersit : 
            - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
            - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
            - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
            - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
        - 11 temp : Normalized temperature in Celsius. The values are divided to 41 (max)
        - 12 atemp: Normalized feeling temperature in Celsius. The values are divided to 50 (max)
        - 13 hum: Normalized humidity. The values are divided to 100 (max)
        - 14 windspeed: Normalized wind speed. The values are divided to 67 (max)
        - 15 casual: count of casual users
        - 16 registered: count of registered users
        - 17 cnt: count of total rental bikes including both casual and registered ()
    '''
    np.set_printoptions(precision=300)
    X = np.genfromtxt("x.csv", delimiter=",")
    Y = np.genfromtxt("y.csv", delimiter=",")
    X_T = np.transpose(X)
    # n, m calculation
    n, m = X_T.shape
    if m <= n:
        print("RECOMENDATION: Delete some features, or use regularization.")

    theta = np.dot(np.dot(np.linalg.pinv(np.dot(X_T, X)), X_T), Y)
    
    y = np.dot(X, theta)
    
    casual_users = X[:, 12] 
    registered_users = X[:, 13]
    total_rental_bikes = Y
    
    X_DF = pd.DataFrame(X)
    # with regression
    

    #print(theta)
    plt.title('Prediction of bike rental on based environmental and seasonal settings')
    plt.ylabel('Total Rental Bikes')
    plt.xlabel('Registered Users')
    plt.plot(y,Y, '-')
    plt.plot(registered_users, total_rental_bikes , '.')
    plt.show()
    # Ploting the scores as scatter plot
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.set_title('Prediction of bike rental based on environmental and seasonal settings')
    ax.set_xlabel('Registered Users')
    ax.set_ylabel('Casual Users')
    ax.set_zlabel('Total Rental Bikes')
    ax.scatter(casual_users, registered_users, total_rental_bikes, color='#ef1234')
    plt.show()
    
if __name__ == '__main__':
    if sys.path[0] == '':
        del sys.path[0]
    main()
```
